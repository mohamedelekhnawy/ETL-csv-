```python
Range_1=csv_1.rename(columns={"Rank2016":"Rank_ID","City served":"City_served","Passengers2015":"Passengers_15","Passengers2016":"Passengers_2016","Change 2016-2015-Num":"difference_between_2015-2016","Change 2016-2015-%":"Change_2016-2015-Persent"})
Range_1.head()
```

```python
Range_2=csv_2.rename(columns={"Rank2017":"Rank_ID","City served":"City_served","Passengers2017":"Passengers_2017","Passengers201":"Passengers_2016","Change 2017–2016-Num":"Change_2017–2016-Num","Change 2017-2016-%":"Change_2017-2016-Persent"})
Range_2.head()
```

```python

colums_drop=Range_1.drop(["Rank_ID","Country","difference_between_2015-2016",'Change_2016-2015-Persent'],axis=1)
colums_drop.head()

# sum two colums Passengers_2016 ,Passenger2015 
df=pd.DataFrame(colums_drop)
df['Passengers_2015']=df['Passengers_2016'] + df['Passengers_15']
df

# # replace two colums 
df.drop(['Passengers_2016', 'Passengers_15'], axis=1, inplace=True)
df
```

```python
colume_drop_2=Range_3.drop(['Rank_ID','Country','difference_between_2017-2018','Change_2018-2017-Persent'],axis=1)
colume_drop_2

#sum columes Passengers_2018 Passengers_2017
df2=pd.DataFrame(colume_drop_2)
df2['Passengers-2018']=df2['Passengers_2018'] + df2['Passengers_2017']
df2

#make drop for Passengers_2018  Passengers_2017 and mke the total onlie
df2.drop(['Passengers_2018','Passengers_2017'],axis=1,inplace=True)
df2
```

```python
Airport_2015_2018=pd.merge(df,df2)
Airport_2015_2018
```

```python
Passengers=pd.DataFrame(Airport_2015_2018)
Passengers['difference_between_2015_2018']=Passengers['Passengers-2018'] - Passengers['Passengers_2015'] 
Passengers
```

```python
Passengers.to_csv('Passengers.csv', index=False)
```

## Creating a table

```python
# Replace these values with your actual database credentials
dbname = 'your_dbname'
user = 'your_user'
password = 'your_password'
host = 'your_host'
port = 'your_port'

# Establish a connection to the database
conn = psycopg2.connect(
    dbname=dbname,
    user=user,
    password=password,
    host=host,
    port=port
)

# Open a cursor to perform database operations
cur = conn.cursor()

# Define the create table query
create_table_query = """
CREATE TABLE IF NOT EXISTS Passengers (
    Airport TEXT,
    City_served TEXT,
    Passengers_2015 INTEGER,
    Passengers_2018 INTEGER,
    Difference_between_2015_2018 INTEGER
);
"""
# Execute the create table query
cur.execute(create_table_query)

# Commit the transaction
conn.commit()

```

```python
import csv 
csv_file = 'Passengers.csv'

# Open the CSV file and insert its contents into the PostgreSQL table
with open(csv_file, 'r') as file:
    reader = csv.reader(file)
    next(reader)  
    for row in reader:
        cur.execute(
            "INSERT INTO Passengers (Airport, City_served, Passengers_2015, Passengers_2018, Difference_between_2015_2018) VALUES (%s, %s, %s, %s, %s)",
            row
        )
        conn.commit()

```
